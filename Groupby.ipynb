## Grouping by Multiple Columns

```python
fortune = pd.read_csv("fortune1000.csv" , index_col = "Rank")
sectors  = fortune.groupby(["Sector" , 'Industry'])
fortune.head(3)
```

```python
sectors.size().head()
sectors.sum()
sectors['Revenue'].sum()
sectors['Employees'].mean()
```

## The agg() Method

```python
sectors.agg({'Revenue':'sum',
            'Profits':'sum',
            'Employees':'mean'}).head()
```

```python
sectors.agg(['size','sum','mean']).head()
```

```python
sectors.agg({'Revenue':['sum','mean'],
            'Profits':'sum',
            'Employees':'mean'}).head()
```

### Iterating through group

```python
fortune.head(2)
```

```python
sectors['Profits'].max().head()
```

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

for  looping through all the dataframe we have to use two temp name one is group and dataframe name that is going to be used

```python
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1 , 'Revenue')
    df  = df.append(highest_revenue_company_in_group)
# data = temparary name
# 1 is used for top value
```

```python
df
```

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for city , data in cities:
    highest_revenue_company_in_city = data.nlargest(1 , 'Revenue')
    df = df.append(highest_revenue_company_in_city)
    
```
